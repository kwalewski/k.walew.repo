### 1.2.Cel projektu

Celem naszego projektu jest zamodelowanie lotu spadochroniarza skaczącego z wysokości 4000ft (1219.2m). Przedstawimy 2 modele opisujące zmiany przyspieszenia, predkości i wysokości w czasie. Porównamy ich dokładność oraz pokażemy ich dobre i złe strony.

## 2. Wstęp do modelowania

Do opisu ruchu skoczka spadochronowego skorzystamy z 2. zasady dynamiki. Na spadochroniarza wraz ze sprzętem działa siła ciężkości skierowana pionowo w dół. W kierunku przeciwnym działa siła oporu. Rozpatrujemy ruch po osi $y$(dla przejrzystości nazwanej $H$), której punkt początkowy znajduje się na powierzchni ziemi. Tym sposobem otrzymujemy następujące równanie:

\begin{equation}a \cdot m = H'' \cdot m =- F_g + F_{d},\:\:\:\:\:\:\:\:\:\:\:\:(1)\end{equation}

gdzie:

   $a$ - przyspieszenie skoczka,
   
   $m$ - masa skoczka i spadochronu,
   
   $H$ - wysokość skoczka nad ziemią,
   
   $F_g$ - siła grawitacji działająca na skoczka,
   
   $F_d$ - siła oporu powietrza działająca na skoczka.
   


Różnica pomiędzy przyspieszeniem grawitacyjnym na ziemi, a na wysokości z której zazwyczaj oddawane są skoki, jest na tyle mała, że możemy uznać ją za stałą:
$g=9.81 \frac{m}{s^2}$.

Z kolei siła oporu w danym ośrodku(u nas w powietrzu)dana jest wzorem $[2]$:

 \begin{equation}F_{d}=k \cdot v^2=\frac{1}{2} \cdot C_x \cdot S \cdot \rho \cdot v^2,\:\:\:\:\:\:\:\:\:\:\:\:(2)\end{equation}  

gdzie:

   $C_x$ - współczynnik siły oporu,
   
   $S$ -  powierzchnia rzutu ciała na płaszczyznę prostopadłą do wektora prędkości ciała,
   
   $\rho$ - gęstość powietrza,
   
   $v$ - prędkość ciała.

Przyjmujemy gęstość powietrza: $\rho = 1 \frac{kg}{m^3}$.

## 3. Modelowanie

### 3.1. Spadek swobodny

Na początek zajmiemy się modelowaniem spadku swobodnego. Da nam to podstawy na których będziemy mogli oprzeć dalsze modele. Zakładamy, że człowiek nie otworzy spadochronu oraz spada równolegle do podłoża(klatką piersiową do dołu). 

### 3.1.1. Równania 



Z równania $(1)$ i $(2)$ mamy:

<center>\begin{equation}H'' \cdot m = -m \cdot g + \frac{1}{2} \cdot C_x \cdot S \cdot \rho \cdot (H')^2.\end{equation}</center> 

Wiemy, że $H'=v$. Po takim podstawieniu dostajemy równanie pierwszego rzędu:

<center>\begin{equation}v' \cdot m = -m \cdot g + \frac{1}{2} \cdot C_x \cdot S \cdot \rho \cdot v^2.\:\:\:\:\:\:\:\:\:\:\:\:(3)\end{equation}</center> 

Przekształcając równanie $(3)$ otrzymujemy:

<center>\begin{equation}v'=\frac{dv}{dt}=-g+\frac{k}{m} \cdot v^2,\end{equation}</center>

teraz całkujemy:

<center>\begin{equation}\int \frac{dv}{-g+\frac{k}{m} \cdot v^2}=\int dt,\end{equation}</center>

zamieniamy na ułamki proste:

\begin{equation}\int -\frac{1}{2} \cdot \sqrt{\frac{m}{k \cdot g}} \left( \frac{1}{\sqrt{\frac{mg}{k}}+v}+\frac{1}{\sqrt{\frac{mg}{k}}-v} \right) dv=\int dt,\end{equation}

dalej:

\begin{equation}-\frac{1}{2}\sqrt{\frac{m}{k \cdot g}}\ln\left|\frac{\sqrt{\frac{mg}{k}}+v}{\sqrt{\frac{mg}{k}}-v}\right|=t+C.\end{equation}

Opuszczamy wartość bezwzględną, ponieważ $v$ nie przekroczy prędkości granicznej (którą zaraz wyliczymy), więc wartość wyrażenia w module będzie dodatnia:

\begin{equation}-\frac{1}{2}\sqrt{\frac{m}{k \cdot g}}\ln\left(\frac{\sqrt{\frac{mg}{k}}+v}{\sqrt{\frac{mg}{k}}-v}\right)=t+C.\end{equation}


Biorąc warunek początkowy $v(0)=0$, wyliczamy stałą $C=0$.

Ostatecznie otrzymujemy:

\begin{equation} v(t)=-\sqrt{\frac{mg}{k}} \frac{e^{\sqrt{\frac{k \cdot g}{m}}t}-e^{-\sqrt{\frac{k \cdot g}{m}}t}}{e^{\sqrt{\frac{k \cdot g}{m}}t}+e^{-\sqrt{\frac{k \cdot g}{m}}t}}=- \sqrt{\frac{mg}{k}} \tanh \left(\sqrt{\frac{k \cdot g}{m}}t\right).\end{equation}

Wiedząc, że $\frac{dH}{dt}=v$ obliczymy $H(t)$:

\begin{equation} \int dH=-\int \sqrt{\frac{mg}{k}} \tanh \left(\sqrt{\frac{k \cdot g}{m}}t\right)dt,\end{equation}

\begin{equation}H(t)= -\frac{\sqrt{\frac{mg}{k}}}{\sqrt{\frac{kg}{m}}} \ln\left({\cosh \sqrt{\frac{kg}{m}}t}\right)+D= -\frac{m}{k} \ln\left({\cosh \sqrt{\frac{kg}{m}}t}\right)+D.\end{equation}

Biorąc warunek początkowy $H(0)=1219.2$ (skok z wysokości 1219.2m), wyliczamy $D=1219.2$:

\begin{equation}H(t)= -\frac{m}{k} \ln\left({\cosh \sqrt{\frac{kg}{m}}t}\right)+1219.2.\end{equation}

Przyjmujemy:

$m=97.2 kg$ - masa skoczka wraz ze sprzętem,

$C_x=1.95$ - współczynnik oporu człowieka, którego przyjęliśmy jako płaską prostokątną płytę $[3]$,

$S=0.5 m^2$ - powierzchnia człowieka w pozycji poziomej $[4]$.

### 3.1.2. Wykresy 

```python
import numpy as np
import math
import matplotlib.pyplot as plt

ts=np.linspace(0,31,num=1000)

m=97.2
g=9.81
S=0.5
Cx=1.95
rho=1
k=0.5*rho*Cx*S

def velocity(t):
    v=-math.sqrt(m*g/k)*math.tanh(math.sqrt(k*g/m)*t)
    return v

def hight(t):
    h=-m/k*math.log(math.cosh(math.sqrt(k*g/m)*t))+1219.2
    return h

H=[]
V=[]
for t in ts:
    V.append(velocity(t))
    H.append(hight(t))

A=[]
for v in V:
    a=-g+k/m*v**2
    A.append(a)

print("wysokość skoczka po 10s lotu: ",hight(10))
print("prędkość skoczka po 10s lotu: ",velocity(10))

plt.figure(figsize=(15,15))   

plt.subplot(3,3,1)
plt.plot(ts,A, color="yellow")
plt.title("przyspieszenie w czasie")
plt.xlabel("czas")
plt.ylabel("przyspieszenie")

plt.subplot(3,3,2)
plt.plot(ts,V, color="blue")
plt.title("prędkość w czasie")
plt.xlabel("czas")
plt.ylabel("prędkość")

plt.subplot(3,3,3)
plt.plot(ts,H, color="black")
plt.title("wysokość w czasie")
plt.xlabel("czas")
plt.ylabel("wysokość")

plt.show()
```

Zwróćmy uwagę, że na skutek działania siły oporu, moduł przyspieszenia skoczka maleje, aż w pewnym czasie osiąga wartość 0. Wtedy prędkość również się stabilizuje na wartości około -44 $\frac{m}{s}$(wektor prędkości jest skierowany w dół, czyli przeciwnie do zwrotu osi $H$, stąd $v<0$). Mówimy, że ciało osiągnęło prędkość graniczną, ponieważ siła oporu równoważy się z siłą ciężkości. Ciało dłużej nie przyspiesza i porusza się ruchem jednostajnym, co widać na trzecim wykresie, gdzie zależność wysokości od czasu staje się liniowa. Widzimy także, że upadek człowieka z 1219 m będzie trwał około 31 sekund.

Dokładna wartość prędkości granicznej:

<center>\begin{equation}F_g=F_{d},\end{equation}</center>

\begin{equation}m \cdot g=k \cdot v^2,\end{equation}

\begin{equation}v_{gr}=-\sqrt{\frac{m \cdot g}{k}}.\end{equation}

Podstawiając dane z zadania otrzymujemy:

<center>\begin{equation}v_{gr}=-44.226 \frac{m}{s} \approx -159,21 \frac{km}{s}.\end{equation}</center>

Dokładny czas spadania:

<center>\begin{equation}H(t)=-\frac{m}{k} \ln\left({cosh \sqrt{\frac{kg}{m}}t}\right)+1219.2=0,\end{equation}</center>
<center>\begin{equation}t = 30,69 s.\end{equation}</center>

### 3.2. Skok ze spadochronem. Model uproszczony.

Teraz zajmiemy się prostym modelowaniem skoku ze spadochronem. W tym celu podzielimy skok na dwa etapy z różnymi oporami powietrza:

### Etap pierwszy 

Skoczek spada swobodnie, tak jak w poprzednim modelu.



### Etap drugi 

W pewnym momencie spadochroniarz pociąga za linkę i otwiera się czasza spadochronu. W tym modelu zakładamy, że dzieje się to natychmiastowo i od razu działa maksymalna siła oporu. Jednocześnie skoczek przechodzi do pozycji wertykalnej (przyjmujemy, że jego opór jest taki sam jak cylindra o wysokości $h$). Czasza spadochronu ma kształt półsfery $[4]$. Wypadkowa siła oporu jest sumą siły oporu spadochronu i człowieka.



### 3.2.1. Równania

Korzystając z  równań $(1)$ i $(2)$ możemy zapisać równanie ruchu skoczka:

\begin{equation}H'' \cdot m=-m\cdot g+\frac{1}{2} \cdot \rho \cdot (H')^2(C_x^b\cdot S^b+C_x^e \cdot S^e),\end{equation}
\begin{equation}v' \cdot m=-m\cdot g+\frac{1}{2} \cdot \rho \cdot v^2(C_x^b\cdot S^b+C_x^e \cdot S^e).\:\:\:\:\:\:\:\:\:\:\:\:(4)\end{equation}

gdzie:

$C_x^b$ - wsp. oporu skoczka

$C_x^e$ - wsp. oporu spadochronu

$S^b$ - pole powierzchni skoczka

$S^e$ - pole powierzchni spadochronu

Załóżmy, że otwarcie spadochronu następuje w chwili $t_0$. Wtedy równanie $(4)$ możemy zapisać jako:

\begin{equation}v' \cdot m=-m\cdot g+k(t)v^2,\end{equation}

gdzie:

<center> $k(t)=\begin{cases}
k_0, \; \; \; \;t \leq t_0 \\
k_1 ,\; \; \; \; t > t_0\end{cases},$   </center>

<br>
<center> $k(t)=\frac{1}{2} \cdot \rho \begin{cases}
C_{x}^{b0} \cdot S^{b0}, \; \;\; \; \; \; \; \;\; \;\;\; \; \;   \; \; \; \; \; \;t \leq t_0 \\
C_{x}^{b1}\cdot S^{b1}+C_{x}^e\cdot S^{e} ,\; \; \; \; t > t_0\end{cases}.$   </center>

Dla $t \leq t_0$ równanie ma takie samo rozwiązanie jak w spadku swobodnym:

\begin{equation} v_0(t)=-\sqrt{\frac{mg}{k_0}} \tanh \left(\sqrt{\frac{k_0 \cdot g}{m}}t\right),\end{equation}

\begin{equation}H_0(t)=-\frac{m}{k} \ln\left({cosh \sqrt{\frac{kg}{m}}t}\right)+1219.2.\end{equation}

Rozwiążemy to równanie dla $t>t_0$ podobnie jak w poprzednim modelu:

\begin{equation}\int \frac{dv_1}{-g+\frac{k_1}{m} \cdot v_1^2}=\int dt,\end{equation}

\begin{equation}-\frac{1}{2}\sqrt{\frac{m}{k_1 \cdot g}}\ln\left|\frac{\sqrt{\frac{mg}{k_1}}+v_1}{\sqrt{\frac{mg}{k_1}}-v_1}\right|=t+C.\end{equation}

Wartość wyrażenia w module w tym przypadku jest ujemna, ponieważ prędkość początkowa jest co do modułu większa od wartości granicznej:

\begin{equation}-\frac{1}{2}\sqrt{\frac{m}{k_1 \cdot g}}\ln\left(\frac{\sqrt{\frac{mg}{k_1}}+v_1}{-\sqrt{\frac{mg}{k_1}}+v_1}\right)=t+C,\end{equation}

\begin{equation}v_1(t)=-\sqrt{\frac{mg}{k_1}} \coth \left(\sqrt{\frac{k_1 \cdot g}{m}}(t+C)\right).\end{equation}

Pomnożenie wartości w logarytmie przez -1 sprawiło, że teraz dostajemy funkcje $v_1$ zależną od cotangensa hiperbolicznego.

Przyjmując $t_0=10$, obliczamy stałą $C$ korzystając z tego, że $v(t_0)=v(10)=-43.19=\lim_{t \to t_0^{+}}v(t)$ (Chcemy zachować ciągłość funkcji $v$):

\begin{equation}C=-9.92236,\end{equation}

\begin{equation}v_1(t)=-\sqrt{\frac{mg}{k_1}} \coth \left(\sqrt{\frac{k_1 \cdot g}{m}}(t-9.92236)\right).\end{equation}

Obliczymy teraz zależność $H_1(t)$:

\begin{equation}\int dH_1=-\int \sqrt{\frac{mg}{k_1}} \coth \left(\sqrt{\frac{k_1 \cdot g}{m}}(t-9.92236)\right)dt,\end{equation}

\begin{equation}H_1(t)= -\frac{m}{k_1} \ln\left({\sinh \sqrt{\frac{k_1\cdot g}{m}}(t-9.92236)}\right)+D.\end{equation}

Po 10 sekundach swobodnego spadku, skoczek jest na wysokości 912.79 m. Funkcja $H_1$ musi być oczywiście ciągła. Dzięki tym informacjom liczymy stałą $D=906.079$:

\begin{equation}H_1(t)= -\frac{m}{k_1} \ln\left({\sinh \sqrt{\frac{k_1 \cdot g}{m}}(t-9.92236)}\right)+906.079.\end{equation}

Dla etapu drugiego przyjmujemy następujące wartości współczynników $[3]$ $[4]$:

$C_x^{b1} = 0.35 \cdot h$, 

$h=1.78$ - wysokość człowieka,

$C_x^e = 1.33$, 

$S^{b1}=0.1$, 

$S^e=43.8$. 

### 3.2.3. Wykresy 

```python
m=97.2
g=9.81
h=1.78
Cxb=0.35*h
Cxe=1.33
Sb=0.1
Se=43.8
rho=1

k0=0.5*rho*1.95*0.5
k1=rho*(Sb*Cxb+Se*Cxe)/2

def velocity0(t):
    v=-math.sqrt(m*g/k0)*math.tanh(math.sqrt(k0*g/m)*t)
    return v

def hight0(t):
    h=-m/k0*math.log(math.cosh(math.sqrt(k0*g/m)*t))+1219.2
    return h

def velocity1(t):
    return -math.sqrt(m*g/k1)*1/(math.tanh(math.sqrt(k1*g/m)*(t-9.92236)))

def hight1(t):
    return -(m/k1)*math.log(math.sinh(math.sqrt(k1*g/m)*(t-9.92236)))+906.079

def acceleration0(v):
    a=-(g/m)+(k0/m)*v**2
    return a
    
def acceleration1(v):
    a=-(g/m)+(k1/m)*v**2
    return a
    
ts0=np.linspace(0,10,num=10000)
ts1=np.linspace(10,170,num=10000)

np.delete(ts1,0)

H0=[]
V0=[]
H1=[]
V1=[]
A0=[]
A1=[]
for t in ts0:
    V0.append(velocity0(t))
    H0.append(hight0(t))

for t in ts1:
    V1.append(velocity1(t))
    H1.append(hight1(t))
    
for v in V0:
    A0.append(acceleration0(v))
    
for v in V1:
    A1.append(acceleration1(v))
    
plt.figure(figsize=(15,15)) 

plt.subplot(3,3,3)
plt.title("wysokość w czasie")
plt.xlabel("czas")
plt.ylabel("wysokość")
plt.plot(ts0,H0, label="etap pierwszy")
plt.plot(ts1,H1, label="etap drugi")
plt.ylim(bottom=0)
plt.legend()

plt.subplot(3,3,2)
plt.title("prędkość w czasie")
plt.xlabel("czas")
plt.ylabel("prędkość")
plt.plot(ts0,V0, label="etap pierwszy")
plt.plot(ts1,V1, label="etap drugi")
plt.legend()


plt.subplot(3,3,1)
plt.title("przyspieszenie w czasie")
plt.xlabel("czas")
plt.ylabel("przyspieszenie")
plt.plot(ts0,A0, label="etap pierwszy")
plt.plot(ts1,A1, label="etap drugi")
plt.legend()

plt.show()
```

Widzimy, że w chwili $t_0$ przyspieszenie skoczka osiąga wartość dodatnią co oznacza, że jego zwrot jest ku górze. Z tego powodu ciało zaczyna znacznie wyhamowywać, co widzimy na wykresie prędkości w czasie. Przyspieszenie dość szybko się zeruje, a prędkość osiąga stałą wartość około $-5.7 \frac{m}{s}$. Również zmiana wysokości w czasie odbywa się dużo wolniej, przez co czas spadku się wydłuża, a skoczek ląduje bezpiecznie na ziemi po około 170 sekundach.

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
plt.title("prędkość w czasie")
plt.xlabel("czas")
plt.ylabel("prędkość")
plt.plot(ts0,V0, label="etap pierwszy")
plt.plot(ts1,V1, label="etap drugi")
plt.xlim(9,12)
plt.legend()

plt.subplot(2,2,2)
plt.title("wysokość w czasie")
plt.xlabel("czas")
plt.ylabel("wysokość")
plt.plot(ts0,H0, label="etap pierwszy")
plt.plot(ts1,H1, label="etap drugi")
plt.legend()
plt.xlim(9,12)
plt.ylim(890,960)
plt.show()
```

Cały proces wyhamowywania odbywa się w ułamkach sekund, przez co zależność $H(t)$ szybko staje się liniowa.

```python
plt.title("przyspieszenie w czasie")
plt.xlabel("czas")
plt.ylabel("przyspieszenie")
plt.plot(ts0,A0, label="etap pierwszy")
plt.plot(ts1,A1, label="etap drugi")
plt.legend()
plt.xlim(8,12)
plt.show()
```

Wartą uwagi jest głębsza analiza przyspieszenia. Przez to, że założyliśmy, że spadochron otwiera się w bardzo krótkim czasie (czas otwierania $=0$), przyspieszenie nie jest ciągłe. Siła oporu działa od razu z pełną mocą (nie ma etapów pośrednich i gładkiego przejścia pomiędzy dwoma etapami). Powoduje to powstawanie ogromnego przyspieszenia ponad 500 $\frac{m}{s^2}$. Przeciążenie na tym poziomie jest w większości przypadków śmiertelne (ciekawostka: w 2003 roku Kenny Bräck podczas rajdu samochodowego przeżył zderzenie z siatką graniczną toru. Przeciążenie jakie wskazały urządzenia wynosiło 214g. Jest to najwyższe odnotowane przeciążenie przeżyte przez człowieka $[7]$). Oczywiste jest, że w rzeczywistości wygląda to inaczej, dlatego spróbujemy ulepszyć nasz model.

### 3.3. Skok ze spadochronem. Model ulepszony


Żeby lepiej zobrazować lot spadochroniarza użyjemy modelu zaproponowanego przez matematyków: Douglasa B. Meade'a i Allana A. Struthersa $[4]$.
Model ten zakłada podzielenie skoku na 5 etapów dla większej dokładności, a przede wszystkim urealnienia wartości przyspieszenia:

### Etap pierwszy

Skoczek spada swobodnie w pozycji poziomej. Ruch ten został już omówiony wcześniej.



### Etap drugi 

Po czasie $t=t_0$ spadochroniarz pociąga za linkę i naciągają się linki od spadochronu. Przyjmujemy opór spadochronu jako opór wydłużającego się cylindra o powierzchni podstawy 0.1$m^2$. Z tego względu wartość siły oporu zmienia się w czasie.



### Etap trzeci 

Po rozwinięciu linek zaczyna się otwierać czasza. Jej kształt przyjmujemy jako półsferę. Jednocześnie skoczek przechodzi z pozycji poziomej do pionowej. Weźmy opór spadochronu jako opór powiększającej się półsfery, a opór skoczka jako opór cylindra. Współczynnik oporu powietrza (spadochronu) jest stały, a siła oporu działająca na niego będzie się zwiększać w czasie wraz ze wzrostem powierzchni półsfery ($S^e(t)=S_{1,2}^e(t)$).



### Etap czwarty 

Po całkowitym rozłożeniu spadochronu $(S=\alpha_1)$, ten chwilowo rozciąga się, a potem z powrotem kurczy$(S=\alpha_1)$. W tym czasie $S^e(t)=S_{2,3}^e(t).$



### Etap piąty 

Powierzchnia spadochronu stabilizuje się. Przyjmujemy opór spadochronu jako opór półsfery. $S^e(t)=\alpha_1$.



### 3.3.1. Równania

W ulepszonym modelu nadal używamy wzoru:

\begin{equation} F_d=\frac{1}{2}C_x \cdot S \cdot \rho \cdot v^2 ,\end{equation}

\begin{equation} F_d=F_d^b+F_d^e=\frac{1}{2} \rho (C_x^b S^b+C_x^e S^e)v^2 .\end{equation}

__Dane o spadochronie(te same co w poprzednim modelu):__

$\alpha_1=43.8m^2$ -powierchnia rozłożonego spadochronu,

$l=8.96m$ - długość linek od spadochronu.

__Dane o spadochroniarzu(te same co w poprzednim modelu):__

$b_0=0.5m^2$ - powierzchnia człowieka w pozycji poziomej,

$1.95$ - wsp. oporu skoczka w pozycji poziomej,

$b_1=0.1m^2$ - powierzchnia człowieka w pozycji pionowej,

$0.35 \cdot h$ - wsp. oporu skoczka w pozycji pionowej(cylindra o wysokości h),

$h=1.78m$ - wysokość człowieka,

$m=97.2kg$ - masa człowieka wraz z ekwipunkiem.

__Czasy po których następują kolejne etapy:__

(dane pochodzące z analizy skoków z wysokości 4000ft przez United States Air Force Academy $[5]$)

$t_0=10s$ - czas pociągnięcia za linkę spadochronu,

$t_1=10.5s$ - czas po którym zaczyna się otwieranie czaszy spadochronu,

$t_2=11.5s$ - czas po którym spadochron się w pełni otworzył i zaczyna się rozciągać,

$t_3=13.2s$ - czas po którym spadochron wraca do normalnych rozmiarów.

__Dzieląc skok na etapy otrzymujemy:__

Powierzchnia ciała skoczka jest wyrażona wzorem:

<center>$S^b(t)=\begin{cases} b_0, \; \; t \leq t_0 \\
b_0, \; \; t_0 < t \leq t_1\\
b_1, \; \; t_1 < t \leq t_2\\
b_1. \; \; t_2 < t \leq t_3\\
b_1, \; \; t \geq t_3 \end{cases}.$  </center>

Współczynnik oporu ciała skoczka jest wyrażony wzorem:

<center> $C_x^b(t)=\begin{cases}
1.95, \; \; \; \; \; \; t \leq t_0 \\
1.95, \; \; \; \; \; \; t_0 < t \leq t_1\\
0.35 \cdot h, \; \; t_1 < t \leq t_2\\
0.35 \cdot h, \; \; t_2 < t \leq t_3\\
0.35 \cdot h, \; \; t \geq t_3 \end{cases}.$   </center>

Powierzchnia spadochronu jest wyrażona wzorem:

<center> $S^e(t)=\begin{cases}
0,   \; \; \; \; \; \; \; \; \;   t \leq t_0 \\
b_1, \; \; \; \; \; \; \; \; t_0 < t \leq t_1\\
S_{1,2}^e(t), \; \; t_1 < t \leq t_2\\
S_{2,3}^e(t), \; \; t_2 < t \leq t_3\\
\alpha_1, \; \; \; \; \; \; \; \; t \geq t_3 \end{cases}.$   </center>

W przedziale czasu od $t_1$ do $t_3$ powierzchnia się zmienia, więc jest opisana pewnymi funkcjami zależnymi od $t$.

Współczynnik oporu spadochronu jest wyrażony wzorem:

<center> $C_x^e(t)=\begin{cases}
0, \; \; \; \; \; \; \; \; \; \; \; \; \; \; \; \;  t \leq t_0 \\
0.35 \cdot l\frac{t-t_0}{t_1-t_0}, \; \; t_0 < t \leq t_1\\
1.33, \; \; \; \; \; \; \; \; \; \; \; \;   t_1 < t \leq t_2\\
1.33, \; \; \; \; \; \; \; \; \; \; \; \;  t_2 < t \leq t_3\\
1.33, \; \; \; \; \; \; \; \; \; \; \; \; t \geq t_3 \end{cases}.$   </center>

Między $t_0$ a $t_1$ współczynnik jest zależny od $t$ (cylinder wydłuża się od wysokości 0 do $l$).


Tak jak w poprzednim modelu równanie $(1)$ opisuje ruch skoczka i możemy je zapisać w następujący sposób:

<center> \begin{equation} m\frac{dv}{dt}=-mg+kv^2. \; \;\; \;\:\:\:\:\:\: (5) \end{equation}
    
Z warunkiem początkowym $v(0)=0$.
    
gdzie:

\begin{equation} k=\frac{1}{2}\rho (C_x^b S^b + C_x^e S^e), \end{equation}
    
dokładniej:

<center> $k(t)=\frac{1}{2} \cdot \rho \begin{cases}
1.95b_0, \; \; t \leq t_0 \\
1.95b_0 +0.35b_1 \cdot l\frac{t-t_0}{t_1-t_0}, \; \; \; t_0 < t \leq t_1\\
0.35b_1\cdot h + 1.33S_{1,2}^e(t), \; \; \; t_1 < t \leq t_2\\
0.35b_1\cdot h + 1.33S_{2,3}^e(t), \; \; \; t_2 < t \leq t_3\\
0.35b_1\cdot h + 1.33\alpha_1 , \; \; \; \; t \geq t_3 \end{cases}.$   </center>

### 3.3.2. Analiza 

Aby nasz model był poprawnie zdefiniowany, równanie $(5)$ musi być różniczkowalne na przedziale $(0,\infty)$. Aby tak było współczynnik $k(t)$, musi być ciągły na $(0,\infty)$. 

$k(t)$ jest ciągła na każdym z pięciu przedziałów (zakładając, że $S_{1,2}^e(t)$ oraz $S_{2,3}^e(t)$ są ciągłe).
Musimy zagwarantować ciągłość na krańcach poszczególnych przedziałów:

\begin{equation} k(t_x)=\lim_{t \to t_x^{+}}k(t)\end{equation} Dla $x \in \{0,1,2,3 \}.$

1.W czasie $t_0$ ciągłość jest widoczna:
\begin{equation} 1,95b_0= 1,95b_0 + 0.35b_1 \cdot l\frac{t_0-t_0}{t_1-t_0}.\end{equation} 

2.W czasie $t_1$:

\begin{equation} 1,95b_0+0.35b_1l=0.35b_1h+1.33S_{1,2}^e(t_1).\:\:\:\:\:\:\:\:\:\:\:\: (6)\end{equation} 

3.W czasie $t_2$:

\begin{equation} S_{1,2}^e(t_2)=S_{2,3}^e(t_2)=\alpha_1.\end{equation}

4.W czasie $t_3$:

\begin{equation} S_{2,3}^e(t_3)=\alpha_1.\end{equation}

Dzięki eksperymentowi $[6]$ wiemy, że powierzchnia spadochronu podczas rozkładania, może być opisana funkcją eksponencjalną:

\begin{equation} S_{1,2}^e(t)=\alpha_0 e^{\beta_0 (t-t_1/t_2-t_1)}.\end{equation}
Przy pomocy warunku $(6)$, możemy wyznaczyć współczynniki $\alpha_0$ i $\beta_0$:
\begin{equation} \alpha_0 = \frac{1.95b_0+0.35b_1(l-h)}{1.33},\:\:\: \beta_0 = \ln \left(\frac{\alpha_1}{\alpha_0}\right).\end{equation}

Z kolei powierzchnia spadochronu podczas rozciągania, a potem kurczenia się do pozycji wyjściowej może być opisana funkcją $[4]$:
\begin{equation} S_{2,3}^e(t)=\alpha_1 \left( 1+\beta_1 sin\left(\pi \frac{t-t_2}{t_3-t_2}\right)\right).\end{equation}

Parametr $\beta_1$ opisuje o ile procent większą powierzchnię uzyskał spadochron w maksymalnym rozszerzeniu. Przyjmujemy, że jest to 115% wartości standardowej, więc $\beta_1=0.15$ .

Równanie opisujące ruch skoczka w tym modelu jest skomplikowane, więc analityczne rozwiązanie go, byłoby żmudne i długotrwałe. Rozwiążemy je więc numerycznie, zakładając że w krótkich przedziałach czasu $\Delta t$ ruch jest jednostajnie przyspieszony.

### 3.3.3. Wykresy

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# Kod podzielimy na 5 etapów każdy odpowiedzialny za jeden z etapów lotu

g = 9.81
rho = 1
a1 = 43.8
b0 = 0.5
b1 = 0.1
h = 1.778
l = 8.96
m = 97.2
t0 = 10
t1 = 10.5
t2 = 11.5
t3 = 13.2
ts1 = np.linspace(0, 10, num=1000)
ts2 = np.linspace(10, 10.5, num=1000)
ts3 = np.linspace(10.5, 11.5, num=1000)
ts4 = np.linspace(11.5, 13.2, num=1000)
ts5 = np.linspace(13.2, 155, num=10000)
v0 = 0
beta1 = 0.15
a0 = (1.95 * b0 + 0.35 * b1 * (l - h)) / 1.33
beta0 = math.log(a1 / a0)
delta_t1 = 10 / len(ts1)
delta_t2 = 0.5 / len(ts2)
delta_t3 = 1 / len(ts3)
delta_t4 = 1.7 / len(ts4)
delta_t5 = 155 / len(ts5)

H0 = 1219.2


# Etap pierwszy

def acceleration1(v):
    a = -g + (0.5 * rho * 1.95 * b0 * v ** 2) / m
    return a


def velocity_and_height1():
    V1 = []
    A1 = []
    H1 = []
    v = 0
    H = H0
    for i in ts1:
        if i == 0:
            v = v0
            H = H0
        elif i <= t0:
            vv = v
            H = H + acceleration1(v) * (delta_t1 ** 2) / 2 + v * delta_t1
            v = v + acceleration1(v) * delta_t1

        V1.append(v)
        A1.append(acceleration1(v))
        H1.append(H)
    return V1, A1, H1


# Etap Drugi
v1a = velocity_and_height1()
v1k = v1a[0][-1]  # prędkość początkowa dla etapu drugiego
H1k = v1a[2][-1]


def acceleration2(v, t):
    a = -g + (0.5 * rho * (1.95 * b0 + 0.35 * b1 * l * (t - t0) / (t1 - t0)) * v ** 2) / m
    return a


def velocity_and_height2():
    V2 = []
    A2 = []
    H2 = []
    v = v1k
    H = H1k
    for i in ts2:
        if i > t0:  # nie bierzemy v z t0 bo funkcja jest otwarta lewostronnie
            vv = v
            H = H + acceleration2(v, i) * (delta_t2 ** 2) / 2 + v * delta_t2
            v = v + acceleration2(v, i) * delta_t2

            V2.append(v)
            A2.append(acceleration2(v, i))
            H2.append(H)
    return V2, A2, H2


# Etap trzeci
v2a = velocity_and_height2()
v2k = v2a[0][-1]  # prędkość początkowa dla etapu trzeciego
H2k = v2a[2][-1]


def acceleration3(v, t):
    a = -g + (0.5 * rho * (0.35 * b1 * h + 1.33 * a0 * math.exp(beta0 * (t - t1) / (t2 - t1))) * v ** 2) / m
    return a


def velocity_and_height3():
    V3 = []
    A3 = []
    H3 = []
    v = v2k
    H = H2k
    for i in ts3:
        if i > t1:
            vv = v
            H = H + acceleration3(v, i) * (delta_t3 ** 2) / 2 + v * delta_t3
            v = v + acceleration3(v, i) * delta_t3

            V3.append(v)
            A3.append(acceleration3(v, i))
            H3.append(H)
    return V3, A3, H3


# Etap czwarty
v3a = velocity_and_height3()
v3k = v3a[0][-1]  # prędkość początkowa dla etapu czwartego
H3k = v3a[2][-1]


def acceleration4(v, t):
    a = -g + (0.5 * rho * (
                0.35 * b1 * h + 1.33 * a1 * (1 + beta1 * math.sin(math.pi * (t - t2) / (t3 - t2)))) * v ** 2) / m
    return a


def velocity_and_height4():
    V4 = []
    A4 = []
    H4 = []
    H = H3k
    v = v3k
    for i in ts4:
        if i > t2:
            vv = v
            H = H + acceleration4(v, i) * (delta_t4 ** 2) / 2 + v * delta_t4
            v = v + acceleration4(v, i) * delta_t4

            V4.append(v)
            A4.append(acceleration4(v, i))
            H4.append(H)
    return V4, A4, H4


# Etap piąty
v4a = velocity_and_height4()
v4k = v4a[0][-1]  # prędkośc początkowa dla etapu piątego
H4k = v4a[2][-1]


def acceleration5(v):
    a = -g + (0.5 * rho * (0.35 * b1 * h + 1.33 * a1) * v ** 2) / m
    return a


def velocity_and_height5():
    V5 = []
    A5 = []
    H5 = []
    H = H4k
    v = v4k
    for i in ts5:
        if i > t3:
            H = H + acceleration5(v) * (delta_t5 ** 2) / 2 + v * delta_t5
            v = v + acceleration5(v) * delta_t5

            V5.append(v)
            A5.append(acceleration5(v))
            H5.append(H)
    return V5, A5, H5


# usuwamy pierwsze elementy z ts'ów żeby punkty się nie powtarzały
ts2 = np.delete(ts2, 0)
ts3 = np.delete(ts3, 0)
ts4 = np.delete(ts4, 0)
ts5 = np.delete(ts5, 0)

plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
plt.plot(ts1, velocity_and_height1()[1], label="etap pierwszy")
plt.plot(ts2, velocity_and_height2()[1], label="etap drugi")
plt.plot(ts3, velocity_and_height3()[1], label="etap trzeci")
plt.plot(ts4, velocity_and_height4()[1], label="etap czwarty")
plt.plot(ts5, velocity_and_height5()[1], label="etap piąty")
plt.legend()
plt.title("Przyspieszenie w czasie")
plt.xlabel("czas")
plt.ylabel("przyspieszenie")

plt.subplot(2, 2, 2)
plt.plot(ts1, velocity_and_height1()[1], label="etap pierwszy")
plt.plot(ts2, velocity_and_height2()[1], label="etap drugi")
plt.plot(ts3, velocity_and_height3()[1], label="etap trzeci")
plt.plot(ts4, velocity_and_height4()[1], label="etap czwarty")
plt.plot(ts5, velocity_and_height5()[1], label="etap piąty")
plt.scatter([10, 10.5, 11.5, 13.2], [velocity_and_height1()[1][-1], velocity_and_height2()[1][-1], velocity_and_height3()[1][-1], velocity_and_height4()[1][-1]],
            color="black")
plt.legend()
plt.title("Przyspieszenie w czasie")
plt.xlabel("czas")
plt.ylabel("przyspieszenie")
plt.xlim(8, 15)
plt.show()
```

Najważniejszą różnicą w stosunku do poprzedniego modelu jest to, że udało nam się uzyskać ciągłość przyspieszenia. Przez rozłożenie całego procesu wyhamowywania w czasie ok. 3 sekund uzyskaliśmy dużo mniejsze przeciążenie, które w rzeczywistości i tak jest jeszcze mniejsze (linki spadochronu się rozciągają, a sam skoczek podpięty jest do urządzenia, które znacznie redukuje szarpnięcie w czasie otwarcia czaszy). Niemniej jednak, jest to wartość nie zagrażająca człowiekowi, nawet gdyby przeciążenie trwało 15 sekund.

Pierwszy z omawianych przez nas modeli jest prosty, co jest jednocześnie jego największą zaletą i wadą. Prostota modelu sprawia że jego równanie da się  stosunkowo łatwo policzyć analitycznie.
